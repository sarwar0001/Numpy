```python
import numpy as np
```

```python
P = np.array([[1,7],[2,1],[3,2]])
print(P)
```

```python
Q = np.array([[1,1],[1,-1],[1,0]])
print(Q)
```

```python
C1 = np.add(P,Q)
print(C1)
```

```python
C2 = np.subtract(P,Q)
print(C2)
```

```python
R = np.array([[1,3,5],[1,2,1]])
print(R)
```

```python
C3 = np.dot(P,R)
print(C3)
```

```python
C4 = np.linalg.det(C3)
print(C3)
```

```python
S = np.array([[1,-1],[2,3]])
C5 = np.linalg.inv(S)
print(C5)
```
